```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("mushrooms.csv")
df=pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
labale = LabelEncoder()
for i in df.columns:
    if df[i].dtype==object or str(df[i].dtype)=='category':
        df[i]=labale.fit_transform(df[i])
df
```

```python
df['class'].value_counts()
```

```python
x=df.drop(columns='class',axis=1)
y=df['class']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0,stratify = y)
```

```python
AdaBoost=AdaBoostClassifier(n_estimators=200, random_state=1)
```

```python
bagging=BaggingClassifier()
```

```python
model=[bagging,AdaBoost]
names=['AdaBoost','bagging']
for name, models in zip(names,model):
    modelfit=models.fit(x_train,y_train)
    y_pre=models.predict(x_test)
    print(f'accuracy score of {name}',accuracy_score(y_test,y_pre))
    print(f'confusion matrix of {name}',confusion_matrix(y_test,y_pre))
    print('\n')
```

```python
importances = AdaBoost.feature_importances_

importance_df = pd.DataFrame({
    'Feature': x_train.columns,
    'Importance': importances
})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)
```
